```python
import pandas as pd
import numpy as np

data = {
    "Name":["Rahul","Priya","Arjun","Neha","Vikram"],
    "Age":[21,22,20,23,21],
    "Course": ["AI","ML","Data Science","AI","ML"],
    "Marks": [85,90,78,88,95]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df["Name"])
print(df[["Name","Marks"]])
print(df.iloc[0])
print(df.loc[2,"Marks"])
```

```python
#Filter data
high_scores = df[df["Marks"] > 80]
print(high_scores)
```

```python
#Adding and Updating columns
#Add Pass/Fail column
df["Result"]=np.where(df["Marks"] >= 80,"Pass","Fail")
```

```python
#Update Neha's marks
df.loc[df["Name"] == "Neha","Marks"] = 92
print(df)
```

```python
#Group by Course and calculate mean
avg_marks = df.groupby("Course")["Marks"].mean()

#Plot bar chart
avg_marks.plot(kind="bar",color="skyblue", edgecolor="black")

plt.title("Average Marks by Course")
plt.xlabel("Course")
plt.ylabel("Average Marks")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],
    "Units_Sold": [120, 135, 150, 200, 220, 210],
    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]
}
 
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Month"],df["Units_Sold"],color='skyblue')
plt.title("Units Sold by Month")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```
