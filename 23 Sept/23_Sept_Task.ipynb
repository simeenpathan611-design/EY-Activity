```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}

df = pd.DataFrame(data)
df
```

```python
#Total revenue per product
total_revenue_per_product = df.groupby('Product')['Revenue'].sum()
print("Total revenue per product:\n",total_revenue_per_product, "\n")

#Average units sold per product
avg_units_sold_per_per_product = df.groupby('Product')['Units_Sold'].mean()
print("Average units sold per product:\n",avg_units_sold_per_per_product)
```

```python
plt.figure(figsize=(8,5))

#Loop each product to plot its monthly trend
for product in df['Product'].unique():
    monthly = df[df['Product'] == product].groupby('Month')['Revenue'].sum()
    plt.plot(monthly.index,monthly.values,marker = 'o',label = product)

plt.title("Monthly Revenue Trend per product")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,4))
total_revenue_per_product.plot(kind='bar',color=['skyblue','orange','green'])
plt.title("Total Revenue per product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.show()
```

```python
plt.figure(figsize=(6,6))
total_revenue_per_product.plot(kind='pie',autopct='%1.1f%%',startangle=140)
plt.title("Contribution of Each Product to Total Revenue")
plt.ylabel("")
plt.show()
```
